```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
ax.set_xlim(0,10)
ax.set_ylim(-5,5)
ax.set_xlabel('x')
ax.set_ylabel(r'$\psi$')

line, = ax.plot(0,0)

k = 2*np.pi
c = 1

dt = .1
# ts = np.arange(0,100, dt)
def animation_frame(ts):
    xs = np.linspace(0,10,100)
    ys = np.cos(k*(xs - c * ts))
    line.set_xdata(xs)
    line.set_ydata(ys)
    return line,
    

animation = FuncAnimation(fig, func=animation_frame, frames=np.arange(0, 100, dt), interval=200)
plt.show()
```
