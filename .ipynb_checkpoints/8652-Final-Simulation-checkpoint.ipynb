```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
```

```python
x = 0.3
θ = .4
```

```python
plt.clf()
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
for i in range(100):
    ids = np.arange(40)
    X_n = np.zeros(40)
    X_n[0] = x
    for i in range(40 - 1):
        u = rand.random_sample()
        X_n[i+1] = θ * X_n[i] + (1 - θ) * (1 if u < X_n[i] else 0)
    ax.plot(ids, X_n)
plt.show()
```

```python
plt.clf()
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
count = np.zeros(20)
for j in range(20):
    for i in range(100):
        ids = np.arange(40)
        X_n = np.zeros(40)
        X_n[0] = x
        for i in range(40 - 1):
            u = rand.random_sample()
            X_n[i+1] = θ * X_n[i] + (1 - θ) * (1 if u < X_n[i] else 0)
        ax.plot(ids, X_n)
        if X_n[35] > 0.01:
            count[j] += 0.01
plt.show()

summ = 0
for i in count:
    summ += i
print(summ/20)
```
