```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def A(δ, ω):
    return 1/np.sqrt((1-ω**2)**2 + (δ*ω)**2)
```

```python
ωs = np.linspace(0,3,100)
As1 = [A(0.1,ω) for ω in ωs]
As2 = [A(1,ω) for ω in ωs]
```

```python
fig, ax = plt.subplots()
ax.plot(ωs,As1, label = "A(0.1,ω)")
ax.plot(ωs,As2, label = "A(1, ω)")
ax.legend()
plt.show()
```
