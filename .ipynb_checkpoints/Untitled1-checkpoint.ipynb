```julia
function factorial(n)
    if (n==0 || n==1)
        return 1
    else
        return n*factorial(n-1)
    end
end
```

```julia
function product(a,n)
    prod = 1;
    for i = 0:n-1
        prod *= (a-i)
    end
    return prod
end
```

```julia
function binomial(a,n)
    return product(a,n)/factorial(n)
end
```

```julia
sum = 0
a = 1/2
for n = 0:50
    term = binomial(a,n)/factorial(n)*(-1)^n
    sum += term
    print(term, '\t', sum, '\n')
end
```
